Write a function to find the longest common prefix string amongst an array of strings.

If there is no common prefix, return an empty string "".



Example 1:

Input: strs = ["flower","flow","flight"]
Output: "fl"
Example 2:

Input: strs = ["dog","racecar","car"]
Output: ""
Explanation: There is no common prefix among the input strings.


```python
class Solution(object):
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        if not strs:
            return ""
        prefix = ""
        for i in range(len(strs[0])): #m
          # print(i)
          for word in strs[1:]: #n-1
            # print(word)
            if i >= len(word) or strs[0][i] != word[i]:
              return prefix
          prefix += strs[0][i]
          # print(prefix)
        return prefix

# sol=Solution()
# print(sol.longestCommonPrefix(["flower","flow","flight"]))
# print(sol.longestCommonPrefix(["dog","racecar","car"]))
```

["flower", "flower", "flower"]  O((n-1) * m) = O(2 * 6) = O(12)
We drop constants and less significant terms because Big-O describes asymptotic behavior.

So:

O((n - 1) × m)
≈ O(n × m)

Because (n - 1) behaves like n as n → ∞.

Space complexity:

“O(m), where m is the length of the longest common prefix.”

💬 And if they ask "Why?"
“Because we’re building and returning a string that holds the prefix, and its size depends on the length of the matching characters across the strings.”

| What You Count          | Space Complexity |
| ----------------------- | ---------------- |
| Just internal variables | **O(1)**         |
| Include result string   | **O(m)**         |


Why we sometimes say O(1) space:
The function only uses a few small things like:

a loop counter

a variable to store the prefix

These don’t grow with the input.

That’s why we say O(1) = constant space.

🚨 But! The result (prefix) grows:
If the common prefix is "flower", it’s 6 characters.

So, we are building a string of size m.

That’s why we also say: O(m) space if we count the result.

✅ What to say in an interview:
“The function uses constant space — O(1) — but the result we return can be up to m characters long, so total space is O(m) if you count the output.”
