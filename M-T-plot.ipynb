```python
import matplotlib.pyplot as plt
import numpy as np

# Data
capacity40_energy = np.array([
    0.22794262054651585, 0.1476551053981843, 0.1295219184739087, 0.1195219184739087,
    0.1055219184739000, 0.1050219184730087, 0.1000219184709087, 0.1000219184039087,
    0.0997593105734364, 0.08877893772059087, 0.08729994156787141
])

capacity40_price = np.array([
    1.1964477981129202, 1.5469124218802663, 1.5432278553770389, 1.5432278553770389,
    1.5432278553770389, 1.5432278553770389, 1.5432278553770389, 1.9432278553770389,
    3.992362169371772, 11.227664594922498, 16.066078828909543
])

# Setup
barWidth = 0.35
ticks = np.arange(len(capacity40_energy))
ticks1 = ticks - barWidth / 2
ticks2 = ticks + barWidth / 2

# Plot with fixed size
fig, ax1 = plt.subplots(figsize=(10, 6))

# Primary bars
ax1.bar(ticks1, capacity40_energy, color='deeppink', width=barWidth,
        edgecolor='deeppink', label="Energy [kWh]")
ax1.set_ylabel(r'$N_{\mathcal{A}}$ [kWh]', fontsize=14, color='deeppink')
ax1.tick_params(axis='y', labelcolor='deeppink')

# Secondary bars (price) on right y-axis
ax2 = ax1.twinx()
ax2.bar(ticks2, capacity40_price, color='gray', width=barWidth,
        edgecolor='gray', alpha=0.7, label="Price [$]")
ax2.set_ylabel('Price [$]', fontsize=14, color='gray')
ax2.tick_params(axis='y', labelcolor='gray')

# X-axis setup
xlabels = ["0/1", "0.1/0.9", "0.2/0.8", "0.3/0.7", "0.4/0.6",
           "0.5/0.5", "0.6/0.4", "0.7/0.3", "0.8/0.2", "0.9/0.1", "1/0"]
ax1.set_xticks(ticks)
ax1.set_xticklabels(xlabels, rotation=45)
ax1.set_xlabel(r'Weight pair ($\gamma_m/\beta_m$)', fontsize=14)

# Layout and grid
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
capacity40_emission = np.array([
    4.7186, 5.3387, 6.0030, 6.5973, 8.2630,
    9.3176, 9.9746, 10.1959, 10.4798, 10.7626, 16.5547
])

capacity40_time = np.array([
    17059.044, 13382.749, 12536.611, 12109.503, 11481.321,
    11112.188, 11013.425, 10981.964, 10994.493, 10927.630, 10826.485
])

# Settings
barWidth = 0.2
ticks = np.arange(len(capacity40_emission))
xlabels = ["0/1", "0.1/0.9", "0.2/0.8", "0.3/0.7", "0.4/0.6", "0.5/0.5",
           "0.6/0.4", "0.7/0.3", "0.8/0.2", "0.9/0.1", "1/0"]

#  Create plot with dual y-axes
fig, ax1 = plt.subplots(figsize=(14, 6))

#  Time bars (left bar of each pair) on secondary y-axis
ax2 = ax1.twinx()
ax2.bar(ticks - barWidth/2, capacity40_time, width=barWidth,
        color='gray', edgecolor='gray', alpha=0.7, label='Time [$s$]')
ax2.set_ylabel(r'$T_{\mathcal{A}}$ [s]', fontsize=16, color='gray')
ax2.tick_params(axis='y', labelcolor='gray')

#  Emission bars (right bar of each pair) on primary y-axis
ax1.bar(ticks + barWidth/2, capacity40_emission, width=barWidth,
        color='deeppink', edgecolor='deeppink', label='Emission [$g$]')
ax1.set_ylabel(r'$M_{\mathcal{A}}$ [g]', fontsize=16, color='deeppink')
ax1.tick_params(axis='y', labelcolor='deeppink')

# X-axis
ax1.set_xticks(ticks)
ax1.set_xticklabels(xlabels, rotation=45, fontsize=12)
ax1.set_xlabel(r'Weight pair ($\alpha_m/\lambda_m$)', fontsize=16)

# Final touches
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
