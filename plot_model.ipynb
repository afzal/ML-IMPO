```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Concatenate

# Two input branches
input_1 = Input(shape=(16,))
input_2 = Input(shape=(8,))

# Simple dense layers
x1 = Dense(32, activation='relu')(input_1)
x2 = Dense(32, activation='relu')(input_2)

# Combine and continue
concat = Concatenate()([x1, x2])
output = Dense(1, activation='sigmoid')(concat)

model = Model(inputs=[input_1, input_2], outputs=output)

# Compile
model.compile(optimizer='adam', loss='binary_crossentropy')

```

```python

# Dummy data
import numpy as np
X1 = np.random.rand(100, 16)
X2 = np.random.rand(100, 8)
y = np.random.randint(0, 2, size=(100,))

# Train
model.fit([X1, X2], y, epochs=5)
```
